1. **Mean**: The mean, also known as the average, is a measure of central tendency calculated by summing up all the values in a dataset and dividing the sum by the total number of values.

2. **Median**: The median is another measure of central tendency that represents the middle value of a dataset when it is arranged in ascending or descending order.

3. **Mode**: The mode is the value that appears most frequently in a dataset.

```python
import random

# Function to generate a random float within a specified range
def generate_random_float(min_value, max_value):
    return round(random.uniform(min_value, max_value), 5)

# Function to generate a row of data
def generate_row():
    crim = random.randint(0, 1)
    zn = generate_random_float(0, 1)
    indus = generate_random_float(0, 10)
    chas = random.randint(0, 1)
    nox = generate_random_float(0, 1)
    rm = generate_random_float(4, 8)
    age = generate_random_float(0, 100)
    dis = generate_random_float(0, 10)
    rad = random.randint(0, 10)
    tax = random.randint(0, 500)
    ptratio = generate_random_float(10, 25)
    b = generate_random_float(0, 1)
    lstat = generate_random_float(0, 10)
    medv = generate_random_float(0, 50)

    return f"{crim},{zn},{indus},{chas},{nox},{rm},{age},{dis},{rad},{tax},{ptratio},{b},{lstat},{medv}\n"

# Generate 100 rows of data
data = "crim,zn,indus,chas,nox,rm,age,dis,rad,tax,ptratio,b,lstat,medv\n"
for _ in range(100):
    data += generate_row()

# Write data to a CSV file
with open("generated_data.csv", "w") as file:
    file.write(data)

print("Data generated successfully.")
```
