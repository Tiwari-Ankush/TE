6.  Turn categorical variables into quantitative variables in Python.

get_dummies() : this method will return dummy variable colums.

concat() : to concatenate dummy columns into DataFrames


```python
dumies = pd.get_dummies(df.author)
```

```python
merged = pd.concat([df,dumies], axis  = 'columns')
merged.drop(['author'],axis = 'columns')
```
